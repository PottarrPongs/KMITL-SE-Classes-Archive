# Problem 1

Implement quick sort where the pivot is chosen randomly

```python
import math, random, statistics
```

```python
def partition(li, lo, hi, p) -> tuple[int, int]:
    mid1, mid2 = lo, hi
    i = lo

    while i <= mid2:
        if li[i] < p:
            li[mid1], li[i] = li[i], li[mid1]
            mid1 += 1
            i += 1

        elif li[i] > p:
            li[i], li[mid2] = li[mid2], li[i]
            mid2 -= 1

        else:
            i += 1

    return mid1, mid2
```

```python
def quicksort_helper(li, lo, hi) -> None:
    if lo < hi:
        pivot = li[random.randint(lo, hi - 1)]
        mid1, mid2 = partition(li, lo, hi, pivot)
        quicksort_helper(li, lo, mid1 - 1)
        quicksort_helper(li, mid2 + 1, hi)
```

```python
def quicksort(A) -> list[int]:
    quicksort_helper(A, 0, len(A) - 1)
    return A
```

```python
assert quicksort([2,7,9,3,4,8,5,1,6])==[1,2,3,4,5,6,7,8,9]
```

# Problem 2

Implement quick select where the pivot is chosen using the median-of-medians method.

Note that the parameter i in quickselect(A,i) is the index in the ascendingly-sorted version of A. Hence,
*   to find the smallest element in A, set i to 0
*   to find the 2nd smallest element in A, set i to 1
*   to find the 3rd smallest element in A, set i to 2
*   ...



```python
def call_mom(li) -> int:
    if len(li) <= 5:
        return statistics.median(li)
    else:
        m = list()
        for i in range(0, len(li), 5):
            sub_list = li[i:i + 5]
            m.append(statistics.median(sub_list))
        return int(statistics.median(m))
```

```python
def quickselect_helper(li, lo, hi, nth) -> int:
    if lo == hi and nth == lo:
        return li[nth]
    else:
        pivot = call_mom(li[lo:hi + 1])
        mid1, mid2 = partition(li, lo, hi, pivot)
        if nth < mid1:
            return quickselect_helper(li, lo, mid1 - 1, nth)
        elif nth > mid2:
            return quickselect_helper(li, mid2 + 1, hi, nth)
        else:
            return li[nth]
```

```python
def quickselect(A,i) -> int:
    result = quickselect_helper(A, 0, len(A) - 1,i)
    return result
```

```python
assert quickselect([2,7,9,3,4,8,5,1,6],2)==3
```

# Problem 3

Implement the divide-and-conquer algorithm that we studied for the closest-pair of points problem.

```python
def distance(p1, p2) -> float:
    return math.hypot(p2[0]-p1[0], p2[1]-p1[1])
```

```python
def closest_pair_helper(Px, Py) -> tuple[list[tuple[float, float]], tuple[tuple[float, float], tuple[float, float]]]:
    n = len(Px)
    if n <= 3:
        best = float("inf")
        pair = None
        for i in range(n):
            for j in range(i+1, n):
                d = distance(Px[i], Px[j])
                if d < best:
                    best = d
                    pair = (Px[i], Px[j])
        return best, pair

    mid = n // 2
    mid_x = Px[mid][0]

    Px_l, Px_r = Px[:mid], Px[mid:]
    Py_l, Py_r = [], []
    for p in Py:
        (Py_l if p[0] <= mid_x else Py_r).append(p)

    d1, pair1 = closest_pair_helper(Px_l, Py_l)
    d2, pair2 = closest_pair_helper(Px_r, Py_r)

    delta, best_pair = (d1, pair1) if d1 < d2 else (d2, pair2)

    strip = [p for p in Py if abs(p[0] - mid_x) < delta]

    for i in range(len(strip)):
        for j in range(i+1, min(i+7, len(strip))):
            d = distance(strip[i], strip[j])
            if d < delta:
                delta = d
                best_pair = (strip[i], strip[j])

    return delta, best_pair
```

```python
def closest_pair(P):
    Px = sorted(P, key=lambda x: x[0])
    Py = sorted(P, key=lambda x: x[1])
    _, pair = closest_pair_helper(Px, Py)
    return pair
```
